```python
# KNN 데이터 시각화, * 학습이 되는건 아니다.

import pandas as pd
df = pd.read_csv("basketball_stat.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
get_ipython().run_line_magic('matplotlib', 'inline')

#스틸과 2점슛 비교
sns.lmplot('STL', '2P', data=df, fit_reg=False,
           scatter_kws={"s":150},
          markers=["o","x"],
          hue="Pos") #hue 는 예측값

plt.title('STL and 2P in 2d plane') #서로 구분하기 어려운 데이터
```

```python
# 어시스트와 2점슛 데이터 시각화
sns.lmplot('AST', '2P', data=df, fit_reg=False,
          scatter_kws={"s":150},
          markers=["o","x"],
          hue="Pos")

plt.title('AST and 2P') #구분하는데 있어서는 의미없는 데이터
```

```python
# 블로킹 3점슛
sns.lmplot('BLK', '3P', data=df, fit_reg=False,
          scatter_kws={"s":150},
          markers=["o","x"],
          hue="Pos")

plt.title('BLK and 3P')
```

```python
#리바운드 3점슛

sns.lmplot('TRB', '3P', data=df, fit_reg=False,
          scatter_kws={"s":150},
          markers=["o","x"],
          hue="Pos")

plt.title('TRB and 3P')
```

# 데이터 다듬기

```python
#분별력 없는 데이터 제거
df.drop(['2P', 'AST', 'STL'], axis=1, inplace = True)
```

```python
df.head()
```

```python
# sklearn의 train_test_split 을 사용하면 라인 한줄로 손쉽게 데이터를 나눌 수 있다.
from sklearn.model_selection import train_test_split

#다음어진 데이터에서 20%를 테스트 데이터로 분류
train, test = train_test_split(df, test_size=0.2) 
```

```python
# 학습 데이터의 갯수를 확인한다, 80개의 데이터가 있다.

train.shape[0]
```

```python
#테스트 데이터의 갯수를 확인 20개

test.shape[0]
```

```python
# 다듬어진 데이터 저장 위치

train.to_csv("basketball_train2.csv", index = None)
test.to_csv("bastketball_test2.csv", index = None)
```
