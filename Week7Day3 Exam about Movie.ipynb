```python
# 나이와 직업은 실제 값이 아니라 그룹을 가리키는 코드 번호
# 데이터셋에 있는 README 파일에 해당 코드 설명이 있다
# 나이와 성별에 따른 어떤 영화의 평균 평점을 계산해보자
# 모든 데이터를 하나의 테이블로 병합하여 계산하면 쉽게 처리 가능
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
# pandas 의 merge 함수를 이용해서 ratings 테이블과 users 테이블 병합하고
# 그 결과를 다시 movies 와 병합
# padas 는 병합하려는 두 테이블에서 중복되는 컬럽 키로 사용
```

```python
data.iloc[0]

# .loc[label] 레이블을 기반으로 인덱싱을 한다.
# pandas 는 순수하게 정수 기반 인덱싱을 위한 일련의 메서드를 제공
# .iloc[position] 위치 정수를 기반으로 인덱싱 한다.
# 여기서 정수 위치 인덱싱은 0에서 축의 길이 -1에 해당된다.

# .iloc 메서드의 특징
# python 인덱싱과 동일하게 0부터 시작한다. (0-based 인덱싱)
# 파이썬과 Numpy 슬라이싱에 거의 가깝게 따른다.
 # 슬라이싱을 사용할때 시작지점은 포함되어 있는 반면 마지막 입력지점은 제외된다.(python 슬라이싱과 유사)
# 불린 배열(boolean array)를 사용하는 것이 가능하다.
# 만약 입력된 인덱스의 범위를 넘어서는 경우에는 idexError 를 발생한다.
 # 슬라이싱을 할때에는 예외적으로 인덱스 범위를 넘어서도 된다
# 비정수 인덱싱을 사용할 경우 indexError 를 일으킨다.
# .loc 과 마찬가지로 뒤에 대괄호[] 를 붙여 사용한다.

# .ioc, .iloc, .ix 메서드의 차이
# .loc 메서드는 레이블을 이용하여 행을 선택하며,
 # 슬라이싱을 할 경우 처음과 끝이 둘 다 포함된다.
# .iloc 는 위치 정수를 이용하여 행을 선택한다.
 # 슬라이싱을 할 경우 처음은 포함되나 끝 요소는 포함이 되지 않는다.
    
# 공통점으로는 둘 다 불린 배열(boolean array) 를 이용한 인덱싱이 가능하다.

#.ix 메서드는 레이블과 위치정수 둘 다 사용이 가능한 메서드
 # 단, 사용하는 것은 지양됨
```

```python
#시리즈 클래스는 NumPy에서 제공하는 1차원 배열과 비슷하지만 
#각 데이터의 의미를 표시하는 인덱스(index)를 붙일 수 있다. 데이터 자체는 값(value)라고 한다.

#시리즈 = 값(value) + 인덱스(index)

#시리즈 생성
#데이터를 리스트나 1차원 배열 형식으로 Series 클래스 생성자에 넣어주면 
#시리즈 클래스 객체를 만들 수 있다. 인덱스의 길이는 데이터의 길이와 같아야 한다. 

#인덱스의 값을 인덱스 라벨(label)이라고도 한다. 
#인덱스 라벨은 문자열 뿐 아니라 날짜, 시간, 정수 등도 가능하다.
```

```python
# 성별에 따른 각 영화의 평균 평점을 구하려면 pivot_table 메서드를 사용한다.
```

```python
mean_ratings = data.pivot_table('rating', index='title',
                                columns='gender', aggfunc='mean')

# rating 이라는 컬럼을 인덱스(title) 컬럼(gender)으로 나누어 표를 만들고
# 표 안에 rating 평균값을 넣어준다

# 여러 분류로 섞인 행 데이터를 열 데이터로 회전시키는 것을 Pivot(피봇)이라고 한다.

# 피봇테이블(pivot table)이란 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 
# 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 말한다.

# Pandas는 피봇테이블을 만들기 위한 pivot 메서드를 제공한다. 
# 첫번째 인수로는 행 인덱스로 사용할 열 이름, 두번째 인수로는 열 인덱스로 사용할 열 이름, 
# 그리고 마지막으로 데이터로 사용할 열 이름을 넣는다.

# Pandas는 지정된 두 열을 각각 행 인덱스와 열 인덱스로 바꾼 후 행 인덱스의 라벨 값이 
# 첫번째 키의 값과 같고 열 인덱스의 라벨 값이 두번째 키의 값과 같은 데이터를 찾아서 해당 칸에 넣는다. 
# 만약 주어진 데이터가 존재하지 않으면 해당 칸에 NaN 값을 넣는다.
```

```python
mean_ratings[:5]
```

```python
# 이렇게 하면 매 로우마다 성별에 따른 평균 영화 평점 정보를 담고 있는 DataFrame 객체가 만들어진다.
# 먼저 250건 이상의 평점 정보가 있는 영화만 추려보자. 데이터를 영하 제목으로 그룹화하고
# size() 함수를 사용해서 제목별 평점 정보 건수를 Series 객체로 얻어낸다.
```

```python
ratings_by_title = data.groupby('title').count()
# count  는 DataFrame 모습으로 전체를 출력해준다.
# non-null 값까지 포함해서 나타내주기 때문에 전체를 출력해서 not-null, 이 필요한 공간에
# 해당 값을 적어주기 위해서 전체로 나타내는 것 같다.

ratings_by_title = data.groupby('title').size()
#  size 는 title을 기준으로 전체 표에서 언급된 수만큼 더해서 출력해준다.


# size() 함수는 Series 객체
```

```python
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
# 250 건 이상인 애들만 따로 뽑아보자
```

```python
active_titles
```

```python
# 250건 이상의 평점 정보가 있는 영화에 대한 색인은 mean_ratings 에서 항목을 선택하기 위해 사용 가능
```

```python
mean_ratings = mean_ratings.loc[active_titles]

# active_titles 리스트 안에 있는 값들에 mean_ratings 을 적용해서 결과값을 mean_ratings 에 할당한다.

# .loc 메서드는 레이블을 이용하여 행을 선택하며,
 # 슬라이싱을 할 경우 처음과 끝이 둘 다 포함된다.
# .loc[label] 레이블을 기반으로 인덱싱을 한다.
```

```python
mean_ratings
```

```python
# 여성에게 높은 평점을 받은 영화 목록을 확인하기 위해 F 컬럼을 내림차순으로 정렬
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
```

```python
top_female_ratings[:10]
```

```python
# 평점 차이 구하기
# 남녀 간의 호불호가 갈리는 영화
# mean_ratings 에 평균 평점의 차이를 담을 수 있는 컬럼을 하나 추가하고, 그 컬럼을 기준으로 정렬
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
# 이제 'diff' 로 정렬하면 여성들이 더 선호나는 영화 순으로 나타난다.
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
```

```python
sorted_by_diff[:10]
```

```python
# 역순으로 정렬한 다음 상위 10개 로우를 잘라내면 남성의 선호도 순으로 확인 가능
```

```python
sorted_by_diff[::-1][:10]
```

```python
# 성별에 관계없이 영화에 대한 호불호가 극명하게 나뉘는 영화를 찾아보자
# 호불호는 평점의 분산이나 표준편차로 측정 할 수 있다.
```

```python
rating_std_by_title = data.groupby('title')['rating'].std() #영화별 표준편차

# 자료의 산포도를 나타내는 수치로, 분산의 양의 제곱근으로 정의된다. 
# 표준편차가 작을수록 평균값에서 변량들의 거리가 가깝다
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles] # active_titles 만 선택
```

```python
rating_std_by_title.sort_values(ascending=False)[:10] # 평점 내림차순으로 Series 정렬

# 남 여 가 서로 준 평점에서 가장 큰 차이가 나는 영화가 dumb & dumber
# DataFrame 객체에 있는 sort_values를 호출하면 해당 변수에 대해 정렬을 할 수 있습니다. 
# ascending 파라미터는 오름차순으로 정렬할지 여부를 결정합니다.
```
